```python
from transformers import pipeline
```

```python
a=pipeline("sentiment-analysis")
```

```python
text=input("Enter the text: ")
print(a(text))
```

```python
gen=pipeline("text-generation")
```

```python
text=input("Enter the text: ")
t=gen(text,max_length=200)
```

```python
print(t[0].get('generated_text'))
```

```python
n=pipeline('ner')
```

```python
text=input("Enter the text: ")
result=n(text)
result
```

```python
qa=pipeline("question-answering")
```

```python
c='''Bitty bought some butter, but the butter was bitter butter.
so, Bitty some better butter, to make bitter butter better'''
```

```python
q='what does Bitty bought?'
```

```python
result=qa({'question':q,'context':c})
print(result)
```

```python
s=pipeline('summarization')
```

```python
t='''Paragraphs can contain many different kinds of information. A paragraph could contain a series of brief examples or a single long illustration of a general point. It might describe a place, character, or process; narrate a series of events; compare or contrast two or more things; classify items into categories; or describe causes and effects. Regardless of the kind of information they contain, all paragraphs share certain characteristics. One of the most important of these is a topic sentence.'''
```

```python
result=s(t)
result[0].get('summary_text')
```
